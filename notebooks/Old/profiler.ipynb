## The large estimate matrices

````python
def plot_estimated_matrices(estimated_matrices_tuple, fig, suptitle, cmap):
    fig.suptitle(suptitle, fontweight = "bold")
    
    axs = fig.subplots(2, 3, height_ratios=[1, 0.1], sharex="col", sharey="row")
    axs = axs.flatten()
    
    axidx = 0
    for item in estimated_matrices_tuple:
        if isinstance(item, dict):
            for subkey, subitem in item.items():
                sns.heatmap(np.log(subitem), ax = axs[axidx], annot = False, cmap = cmap)
                axs[axidx].set_title(r"$\hat{F}_{" + str(subkey) + r"}^{I, J}$, rank " + str(np.linalg.matrix_rank(subitem)))
                
                # ticklabels = axs[axidx].get_yticklabels()
                axs[axidx].set_xticklabels([], rotation = 0)
                axs[axidx].set_yticklabels([], rotation = 0)
                
                axidx += 1
        else:
            if item.shape[0] > item.shape[1]:
                item = item.T
                axs[axidx].set_title("Column Vector (${}^T$)")
            else:
                axs[axidx].set_title("Row Vector")
            sns.heatmap(np.log(item), ax = axs[axidx], annot = False, cmap = cmap)

            # ticklabels = axs[axidx].get_yticklabels()
            axs[axidx].set_xticklabels([], rotation = 0)
            axs[axidx].set_yticklabels([], rotation = 0)
            
            axidx += 1
    
    for axidx_inv in range(axidx, len(axs)):
        axs[axidx].set_visible(False)
````

````python
fig = plt.figure(figsize=(12, 9), constrained_layout = True)
subfigs = fig.subfigures(2, 1)
subfigs = subfigs.flatten()

N = len(estimated_matrices_all)
cmap_names = ["rocket", "mako"]
which_mats = [0, -1]

for idx, subfig in enumerate(subfigs):
    plot_estimated_matrices(
        estimated_matrices_all[which_mats[idx]],
        subfig,
        suptitle = f"Estimated matrices for $L = 10^{3 if idx==0 else 5}$ at $L_c = L_i = 2$",
        cmap = sns.color_palette(cmap_names[idx % len(cmap_names)], as_cmap = True)
    )
fig.savefig(f"Matrices_{source_name}.pdf")
````

## The properties of a sequence

````python
from collections import Counter

from matplotlib import pyplot as plt
from matplotlib import cm, colors
import seaborn as sns
import pandas as pd
import numpy as np

from src.experiments.util_experiments import make_source
from src.oom import ContinuousValuedOOM

np.set_printoptions(linewidth=200)

import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from src import oom
from src.experiments.experiment_blended_known import experiment_continuous_dataset, estimate_matrices_continuous
from src.oom.util.learning_continuous import process_matrices_continuous
from src.experiments.util_experiments import *
from src.oom.discrete_observable import *

sys.modules['oom'] = oom
````

````python
mfs = get_gaussian(seed = 100)(8)
source_name = "OCT8_L10_D100"
gen_length = 10000
binsd2g = ContinuousValuedOOM.from_discrete_valued_oom(make_source(source_name), mfs)
gen = binsd2g.generate(gen_length, reduced=False)
````

````python
def plot_memberships(gen, oom, fig, suptitle):
    fig.suptitle(suptitle, fontweight = "bold")
    axs = fig.subplots(1, 3, width_ratios=[0.5, 1, 1])
    axs = axs.flatten()
    
    x = np.linspace(-4, 4, 8000)
    
    axs[0].set_title("Sampled MF (%)")
    ser = pd.Series([obs.uid for obs in gen.sequence])
    ser = ser.value_counts(normalize = True)
    ser = ser.sort_index(key = lambda x: x.str.ljust(5, 'z'))
    axs[0].bar(x = range(1, len(ser.index) + 1), height = ser, alpha=0.7)
    axs[0].set_xticks(ticks = range(1, len(ser.index) + 1), labels = list(ser.index))
    
    axs[1].set_title("Membership functions")
    for idx, mf in enumerate(oom.membership_fns):
        axs[1].plot(x, mf.pdf(x), label=r"$\nu_{" + oom.observables[idx].uid + r"}$", lw=1)
    axs[1].plot(x, np.average([mf.pdf(x) for mf in oom.membership_fns], weights=ser.values, axis=0),
                c='r', ls='--', label="MF sum")
    
    axs[2].set_title("Empirical distribution (vs. weighted MF sum)")
    axs[2].hist(gen.sequence_cont, bins=60, density=True, alpha=0.7)
    for idx, mf in enumerate(oom.membership_fns):
        axs[2].plot(x, mf.pdf(x), alpha=0)
    axs[2].plot(x, np.average([mf.pdf(x) for mf in oom.membership_fns], weights=ser.values, axis=0),
                c='r', ls='--')
    
    handles, labels = axs[1].get_legend_handles_labels()
    axs[2].legend(handles, labels)


def plot_sequences(gen, oom, smoothing_n, fig, suptitle):
    fig.suptitle(suptitle, fontweight = "bold")
    axs = fig.subplots(3, 2, sharex = 'col', width_ratios = [1, 1e-2])
    axs = axs.flatten()

    axs[0].set_title(f"Probability of each observable (smoothed over {smoothing_n} steps)")
    for obsidx, obs in enumerate(oom.observables):
        axs[0].plot(
            np.convolve(
                [v[obsidx] for v in gen.p_vec_list],
                np.ones(smoothing_n) / smoothing_n,
                mode = "valid"
            ),
            label = f"$P({obs.uid})$"
        )
    axs[0].set_ylim([0, 1])
    axs[0].set_xlim([0, len(gen.p_vec_list) - smoothing_n])
    axs[0].legend()

    axs[2].set_title(f"Sampled membership function (smoothed over {smoothing_n} steps)")
    axs[2].plot(
        np.convolve(
            [int(oom.observables.index(obs)) for obs in gen.sequence],
            np.ones(smoothing_n) / smoothing_n,
            mode = "valid"
        )
    )
    axs[2].set_yticks(list(range(len(oom.observables))), labels = [obs.uid for obs in oom.observables])
    axs[2].set_ylim([0, len(oom.observables) - 1])
    axs[2].set_xlim([0, len(gen.sequence) - smoothing_n])

    axs[4].set_title(f"Magnitude of each dimension in the states (smoothed over {smoothing_n} steps)")
    cmap = sns.color_palette("viridis", as_cmap=True)
    for dim in range(oom.dim):
        axs[4].plot(
            np.convolve(
                [v[dim].item() for v in gen.state_list],
                np.ones(smoothing_n) / smoothing_n,
                mode = "valid"
            ),
            c = cmap( dim / (oom.dim - 1) ),
            label = r"$\omega_" + str(dim + 1) + "$"
        )
    axs[4].set_xlim([0, len(gen.state_list) - smoothing_n])
    
    axs[1].set_visible(False)
    axs[3].set_visible(False)
    fig.colorbar(
        cm.ScalarMappable(norm = colors.Normalize(1, oom.dim), cmap = cmap),
        cax = axs[5],
        orientation = 'vertical', fraction = 0.01, pad = 0.01,
        ticks = list(range(1, oom.dim + 1)), label = 'Dimension'
    )
    
    fig.supxlabel("Sequence time step")
````

````python
fig = plt.figure(figsize = (12, 13), layout='compressed')
subfigs = fig.subfigures(2, 1, height_ratios = [1, 3])
subfigs = subfigs.flatten()

plot_memberships(
    gen,
    oom = binsd2g,
    fig = subfigs[0],
    suptitle = f"Properties of memberships in a {source_name} sequence ($L_g = {gen_length}$)"
)
plot_sequences(
    gen,
    binsd2g,
    smoothing_n = 100,
    fig = subfigs[1],
    suptitle = f"Properties of generation in a {source_name} sequence ($L_g = {gen_length}$)"
)
fig.savefig(f"Properties_{source_name}.pdf")
````
