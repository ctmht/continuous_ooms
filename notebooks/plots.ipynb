```python
from matplotlib import pyplot as plt
from matplotlib import colors as mcolors
from matplotlib import cm
import matplotlib as mpl
import seaborn as sns
import pandas as pd
```

```python
dfd = pd.read_csv('data/experiment_results/experiment_discrete_dataset_results_1.csv')
```

```python
per_graph_nlines = 5

for val in dfd['name'].unique():
    df1 = dfd[dfd['name'] == val]
    title1 = f"Learning performance for {val}"

    nlines = len(df1['target_dimension'].unique())
    palette = sns.color_palette("husl", nlines)

    if nlines > per_graph_nlines:
        fig, axs = plt.subplots(nlines // per_graph_nlines + 1, 1, figsize=(8, 8),\
                                sharex=True, sharey=True, gridspec_kw={'hspace': 0})
        axs = axs.flatten()
    else:
        fig = plt.figure()
        ax = fig.gca()

    # fig.set_layout_engine('constrained')
    fig.suptitle(t = title1 + title2, y = 0.95)
    fig.supxlabel('Generation length $L_g$')
    fig.supylabel('Estimated entropy rate of learned OOM')
    
    for idx, dim in enumerate(df1['target_dimension'].unique()):
        title2 = f" at $d={dim}$"
        if idx % per_graph_nlines == 0:
            if nlines > per_graph_nlines:
                ax = axs[idx // per_graph_nlines]
                
        df2 = df1[df1['target_dimension'] == dim]
        ax.plot(df2['length'], df2['nll_learned'], label=f"{dim=}", color=palette[idx])
        ax.set_xlim([df1['length'].min(), df1['length'].max()])
        ax.set_xscale('log')
        ax.legend(loc='center', bbox_to_anchor=(1.1, 0.5))
        ax.grid(alpha=0.5, zorder=-3)
```
